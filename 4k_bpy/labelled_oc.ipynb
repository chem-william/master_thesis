```python
interpolation_length = 2500
def load_data(traces, labels):
    traces = np.log10(traces)

    worked_traces = []
    raw_traces = []
    hists = []
    new_labels = []
    for trace, label in tqdm(zip(traces, labels), total=len(traces)):
        if len(trace) == 0:
            continue

        if label == -1:
            continue
        
        raw_trace = trace[trace < PLT_HIGH]
        raw_trace = truncate_trace(raw_trace, threshold=PLT_LOW)
        raw_traces.append(raw_trace)

        thresh = np.argmax(trace <= HIGH)
        trace = trace[thresh:]
        trace = truncate_trace(trace, threshold=LOW)
        worked_traces.append(trace)

        hist, _ = np.histogram(
            trace, bins=BINS, range=(LOW, HIGH)
        )
        hists.append(hist)
        new_labels.append(label)

    hists_2d = create_2dhists(worked_traces, bins=(128, 128))

    worked_traces = np.array(worked_traces)
    raw_traces = np.array(raw_traces)
    hists = np.array(hists)
    hists_2d = np.array(hists_2d)
    new_labels = np.array(new_labels)

    return hists_2d, hists, worked_traces, raw_traces, new_labels

def create_2dhists(traces, bins):
    hists_2d = []
    for trace in tqdm(traces):
        H, *_ = np.histogram2d(
            trace,
            np.arange(len(trace)),
            bins=bins,
            range=[[LOW, HIGH], [0, 1024 + 256]],
        )
        hists_2d.append(H)
    
    hists_2d = np.array(hists_2d)

    return hists_2d

def pad_traces(traces):
    padded = []
    for trace in traces:
        tmp_trace = np.zeros(1510)
        tmp_trace[:len(trace)] = trace
        padded.append(tmp_trace)
    padded = np.array(padded)
    return padded

def plot_data(traces, labels, vmin, vmax, ax=None, plot_2d=True, use_labels=True, show_original=True):
    if ax is None:
        fig, ax = plt.subplots()

    counts, binedges = np.histogram(traces, bins=100, range=(-7, -0.5))
    binedges = (binedges[1:] + binedges[:-1])/2
    ax.plot(
        binedges,
        counts/len(traces),
        label=f"Original" if show_original else "",
        c="k",
        linestyle="--",
        alpha=0.8,
    )

    for label in np.unique(labels):
        plot_traces = traces[labels == label]
        counts, binedges = np.histogram(plot_traces, bins=100, range=(-7, -0.5))
        binedges = (binedges[1:] + binedges[:-1])/2
        ax.plot(
            binedges,
            counts/len(traces),
            label=f"{label}: {len(plot_traces)}" if use_labels else "",
            c=c[int(label) % len(c)]
        )
        if use_labels:
            ax.legend(frameon=False)

    if plot_2d:
        fig, ax = plt.subplots(1, len(np.unique(labels)))
        for idx, label in enumerate(np.unique(labels)):
            plot_traces = traces[labels == label]
            xaxis = np.concatenate([np.arange(len(x)) for x in plot_traces])
            plot_traces = np.concatenate(plot_traces)
            ymin, ymax = -7, 0
            xmin, xmax = 0, 2500
            bins = 256
            H, *_ = np.histogram2d(
                plot_traces,
                xaxis,
                range=[[ymin, ymax], [xmin, xmax]],
                bins=bins,
            )
            
            im = ax[idx].imshow(
                H,
                origin="lower",
                vmin=vmin,
                vmax=vmax,
            )
        fig.colorbar(im)
        plt.show()

def low_aligned_plot(traces, labels, vmin, vmax):
    low_aligned = []
    for trace in traces:
        thresh = np.argmax(trace < -5)
        trace = trace[thresh - 1000:thresh + 400]
        low_aligned.append(trace)
    low_aligned = np.array(low_aligned)

    fig, ax = plt.subplots(1, len(np.unique(labels)))
    for idx, label in enumerate(np.unique(labels)):
        plot_traces = traces[labels == label]
        xaxis = np.concatenate([np.arange(len(x)) for x in plot_traces])
        plot_traces = np.concatenate(plot_traces)
        ymin, ymax = -7, 0
        xmin, xmax = 0, 1400
        bins = 256
        H, *_ = np.histogram2d(
            plot_traces,
            xaxis,
            range=[[ymin, ymax], [xmin, xmax]],
            bins=bins,
        )
        
        im = ax[idx].imshow(
            H,
            origin="lower",
            vmin=vmin,
            vmax=vmax,
        )
    # fig.colorbar(im)
    plt.show()


def plot_lengths(traces, labels, cutoffs=[-7, -.5]):
    truncated_traces = []
    new_labels = []
    for trace, label in tqdm(zip(traces, labels)):
        top = np.argmax(trace < cutoffs[1])
        trace = trace[top:]
        trace = truncate_trace(trace, threshold=cutoffs[0])
        
        # if len(trace) < 500:
        #     continue
        truncated_traces.append(trace)
        new_labels.append(label)
    truncated_traces = np.array(truncated_traces)

    for label in np.unique(labels):
        class_traces = truncated_traces[new_labels == label]
        
        lengths = np.array([len(x) for x in class_traces])

        counts, binedges = np.histogram(lengths, bins=100, range=(0, 2200))
        binedges = (binedges[1:] + binedges[:-1])/2
        plt.plot(binedges, counts/len(class_traces), c=c[int(label)])
        mean = np.mean(lengths)
        plt.axvline(mean, linestyle="--", c=c[int(label)], label=f"Mean: {mean:.5}")
    plt.legend(frameon=False)
    plt.show()
```

```python
from sklearn import metrics
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, random_state=RNG_STATE, shuffle=True)

# crossvalidate the performance of each model

avg_fpr, avg_tpr = [], []
auc = []
mean_fpr = np.linspace(0, 1, 100)
for model in tqdm(models):
    avg_acc, avg_roc_auc, avg_fm = [], [], []
    tmp_tpr = []
    for idx, (train_indices, test_indices) in enumerate(kfold.split(hists)):
        train_labels = labels[train_indices]
        test_labels = labels[test_indices]
        train_subset = hists[train_indices][train_labels == 1]
        test = hists[test_indices]

        model.fit(train_subset)

        pred_labels = model.predict(test)

        pred_scores = model.decision_function(test)

        if hasattr(model, "kernel"):
            if model.kernel == "linear":
                test_labels *= -1

        if model.__class__.__name__ in ["AutoEncoder", "VAE", "DeepSVDD"]:
            test_labels[test_labels == 1] = 0
            test_labels[test_labels == -1] = 1
        
        avg_acc.append(
            metrics.accuracy_score(
                test_labels,
                pred_labels,
            )
        )
        avg_fm.append(
            metrics.fowlkes_mallows_score(
                test_labels,
                pred_labels,
            )
        )
        if model.__class__.__name__ in ["GMMOutlierDetector"]:
            test_labels *= -1
        avg_roc_auc.append(
            metrics.roc_auc_score(
                test_labels,
                pred_scores,
            )
        )

        fpr, tpr, _ = metrics.roc_curve(test_labels, pred_scores)
        interp_tpr = np.interp(mean_fpr, fpr, tpr)
        interp_tpr[0] = 0.0
        tmp_tpr.append(interp_tpr)
    avg_tpr.append(tmp_tpr)
    auc.append(np.mean(avg_roc_auc))
    tmp_tpr = []

    print(f"Average accuracy for {model.__class__.__name__}: {np.mean(avg_acc)*100:.3} ± {np.std(avg_acc, ddof=1)*100:.3}")
    print(f"Average Fowlkes-Mallows for {model.__class__.__name__}: {np.mean(avg_fm)*100:.3} ± {np.std(avg_fm, ddof=1)*100:.3}")
    print(f"Average ROC AUC for {model.__class__.__name__}: {np.mean(avg_roc_auc)*100:.3} ± {np.std(avg_roc_auc, ddof=1)*100:.3}")
    print("="*25)
```
